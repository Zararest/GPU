# Исследование умнжения матриц

Здесь представлены графики зависимости времени умножения двух матриц 2048xN и Nx2048 от N.

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib as matplotlib
import numpy as np
import Boot as boot
```

```python
N_array = np.linspace(100, 4096 * 2, 7)
fig, ax = plt.subplots(figsize=(10, 7))

time_array_tiled = boot.run_matmul('./build/MatMul', '--tiled', N_array)
time_array_tiled_int = boot.run_matmul('./build/MatMul', '--tiled-integer', N_array)
time_array_simple = boot.run_matmul('./build/MatMul', '--simple', N_array)

ax.plot(N_array, time_array_tiled)  
ax.scatter(N_array, time_array_tiled, marker='+', label='с shared памятью')

ax.plot(N_array, time_array_simple)  
ax.scatter(N_array, time_array_simple, marker='o', label='простой алгоритм')

ax.plot(N_array, time_array_tiled_int)  
ax.scatter(N_array, time_array_tiled_int, marker='x', label='с shared памятью на int')

ax.set_xlabel('Размер смежных сторон')
ax.set_ylabel('Время умножения[мс]')
ax.set_title('Умножение матриц')
ax.legend()
ax.grid()
plt.show()
```
